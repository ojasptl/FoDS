```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df.isnull().sum()
```


### Handling Missing Values

Since there are no missing values in the dataset, as confirmed by the previous cell, we can proceed to the next step: data transformation.


```python
df.dtypes
```

```python
df_ZScaled = pd.DataFrame((df - df.mean()) / df.std())
df_ZScaled
```

```python
plt.plot(df_ZScaled)
plt.title('Z-Score Normalized Data')
plt.show()
```

```python
plt.plot(df)
plt.title('Original Data')
plt.show()
```


### Comparison of Original and Z-Score Normalized Data

In the previous cells, we plotted two graphs to visualize the original and Z-score normalized data from the heart disease dataset. Below is an explanation of these graphs and any abnormalities observed.

#### Original Data Plot
The first plot represents the original data without any scaling or normalization. Each line in the plot corresponds to a feature in the dataset. The y-axis represents the raw values of these features.

- **Observation**: The features have different ranges and scales. For example, the `age` feature ranges from around 30 to 80, while the `chol` (cholesterol) feature ranges from around 100 to 400. This discrepancy in scales can affect the performance of machine learning models, as features with larger ranges can dominate those with smaller ranges.

#### Z-Score Normalized Data Plot
The second plot represents the Z-score normalized data. Z-score normalization scales the data such that each feature has a mean of 0 and a standard deviation of 1. This is achieved by subtracting the mean and dividing by the standard deviation for each feature.

- **Observation**: After normalization, all features are on a similar scale, with most values falling between -3 and 3. This makes it easier to compare the features and ensures that no single feature dominates the others in the analysis or modeling process.

#### Abnormalities
- **Original Data**: The original data plot shows significant variability in the scales of different features. This can lead to issues in machine learning models, as features with larger scales can disproportionately influence the model's performance.
- **Z-Score Normalized Data**: The normalized data plot shows that all features are now on a similar scale, which is desirable for most machine learning algorithms. However, any outliers in the data will still be present, but their influence will be reduced due to the normalization process.

By normalizing the data, we ensure that all features contribute equally to the analysis, improving the performance and reliability of machine learning models.


## Data Covariance and Correlation

In this notebook, we explore the covariance and correlation of the Iris dataset. Covariance and correlation are two statistical measures used to understand the relationship between variables.

### Correlation
Correlation measures the strength and direction of a linear relationship between two variables. It is a dimensionless number that ranges from -1 to 1. A correlation of 1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear relationship, and 0 indicates no linear relationship.

We use a heatmap to visualize the correlation matrix of the Iris dataset. The heatmap provides a color-coded matrix where each cell represents the correlation between two variables. The color intensity indicates the strength of the correlation.

### Covariance
Covariance measures the joint variability of two random variables. Unlike correlation, covariance is not dimensionless and its value depends on the units of the variables. A positive covariance indicates that the variables tend to increase together, while a negative covariance indicates that one variable tends to increase when the other decreases.

We also use a heatmap to visualize the covariance matrix of the Iris dataset. Similar to the correlation heatmap, the covariance heatmap provides a color-coded matrix where each cell represents the covariance between two variables.

### Visualizations
- **Correlation Heatmap**: This heatmap shows the correlation between the features of the Iris dataset. It helps in identifying which features are strongly correlated with each other.
- **Covariance Heatmap**: This heatmap shows the covariance between the features of the Iris dataset. It helps in understanding the joint variability of the features.

The following cells in the notebook contain the code to generate these heatmaps.


```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
x, y = load_iris(return_X_y=True)
df = pd.DataFrame(x)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df['Species'] = y
df
```

```python
df
```

## Correlation


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.corr(), annot= True, cmap = 'Blues', fmt = '.2f')
plt.title("Heatmap of Correlation")
plt.show()
```

## Covariation


```python
from sklearn import covariance
crr = covariance.ShrunkCovariance()
crr.fit(df.drop(columns='Species'), df['Species'])
```

```python
sns.heatmap(crr.covariance_, annot= True, cmap = 'Purples', fmt = '.2f')
plt.title("Heatmap of Covariance")
plt.show()
```

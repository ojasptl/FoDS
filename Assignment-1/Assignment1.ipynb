```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import os
def download_file(url, filename):
    if not os.path.isfile(filename):
        os.system(f'wget {url} -O {filename}')
```

```python
download_file('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv', 'titanic.csv')
df_csv = pd.read_csv('titanic.csv')
print("CSV Data:")
print(df_csv.head())
```

```python
df_excel = pd.read_excel('/content/Sachdeva Store Data Analysis.xlsx')
print("\nExcel Data:")
print(df_excel.head())
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=df_csv)
plt.title('Survival Count by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_excel['Amount'])
plt.title('Sales Distribution')
plt.show()
```
